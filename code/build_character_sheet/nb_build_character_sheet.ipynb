```python
t= pd.DataFrame(df_talents_raw.iloc[0, 0])
t.groupby(['tree','name','type','rank','requires'])['effect'].apply('<br>'.join).reset_index()
```

```python
df_talents = pd.DataFrame()
for i in df_talents_raw.index:
    t = pd.DataFrame(df_talents_raw.iloc[i, 0])
    v = t.groupby(['tree','name','type','rank','requires'])['effect'].apply('<br>'.join).reset_index()
    df_talents = pd.concat([df_talents, v], axis=0)
df_talents.reset_index(drop=True, inplace=True)
df_talents['name'] = df_talents['name'].astype('str')
```

```python
df_talents.head()
```

```python
skills = ["Exert","Grapple","Intimidate","Martial Arts","Shoot","Strike","Throw","Bellor","Inferno","Sangui","Tempest","Implements","Insight","Legerdemain","Operate","Stealth","Tactics","Tumble","Mori","Nox","Plaga","Squall","Craft","Pilot","Analyze","Computers","Cosmology","Current Events","Deceive","History","Linguistics","Nature","Navigate","Persuade","Research","Skrimisiology","Streetwise","Survival","Theology","Arithm","Medicine","Mentis","Phylla","Rime","Torrent","Composure","Endure","Focus","Investigate","Lead","Perceive","Forma","Locus","Lux","Ordi","Tellus","Aether","Destroy","Intuit","Luck","Nihil","Resist","Tempor","Vivus"]
```

```python
for s in skills:
    df_talents[s] = [s if s in eff else None for eff in df_talents['effect']]
    df_talents[s].dropna(inplace=True)
```

```python
def build_powercard(name, effect, skill, type):
    powercard = '''!power {{{{
    --name|@{{name}} - ''' + name + '''
    --Effect|''' + effect + '''
    --Result Set|[[ [$talent|XPND] @{{BAMF|challenge}}d@{{skill_''' + skill.lower() + '''}}>4]]
    --Hits|[^talent.ss]
    --1s|[^talent.ones]
    --format|''' + type.lower() + '''check
    }}}}'''
    return(powercard)
```
