```python
def build_powercard(name, effect, skill, type):
    skill_clean = skill.lower()
    skill_clean = skill_clean.replace(' ', '_')
    powercard = '''!power {{
    --name|@{name} - ''' + name + '''
    --Effect|''' + effect + '''
    --Result Set|[[ [$talent|XPND] @{BAMF|challenge}d@{skill_''' + skill_clean + '''}>4]]
    --Hits|[^talent.ss]
    --1s|[^talent.ones]
    --format|''' + type.lower() + '''check
    }}'''
    return(powercard)
```

#### Build Tech Powercards

```python
tech_pcards = []
for index, t in df_techniques.iterrows():
    skill_clean = t['skill'].lower()
    res = '''!power {{
--Name|@{name} - ''' + t['name'] + '''
--Cost|''' + str(t['cost']) + '''
--Current|''' + str(t['current']) + '''
--api_setattr|_name @{name} _current|''' + str(t['current']) + '''
--Result Set|[[ [$tech|XPND] @{BAMF|challenge}d@{skill_''' + skill_clean + '''}>@{BAMF|target}]]
--Hits|[^tech.ss]
--1s|[^tech.ones]
--format|powercheck\n'''
    if len(t['threshold']) > 1:
        for i in range(len(t['threshold'])):
            if i < len(t['threshold']) - 1:
                clean_effect = t['threshold'][i]['effect'].replace('choose', '**choose').replace(':<ul><li>', '**:^^^^**').replace('</li><li>', '^^^^').replace('</li></ul>', '**')
                res += '''--?? $tech.ss >= ''' + str(t['threshold'][i]['hits']) + ''' AND $tech.ss < ''' + str(t['threshold'][i+1]['hits']) + ''' ?? Effect:|**//Threshold - [^tech.ss]//**^^''' + clean_effect + '\n'
            else:
                clean_effect = t['threshold'][i]['effect'].replace('choose', '**choose').replace(':<ul><li>', '**:^^^^**').replace('</li><li>', '^^^^').replace('</li></ul>', '**')
                res += '''--?? $tech.ss >= ''' + str(t['threshold'][i]['hits']) + ''' ?? Effect:|**//Threshold - [^tech.ss]//**^^''' + clean_effect + '\n'
    else:
        for r in t['threshold']:
            clean_effect = r['effect'].replace('choose', '**choose').replace(':<ul><li>', '**:^^^^**').replace('</li><li>', '^^^^').replace('</li></ul>', '**')
            res += '''--?? $tech.ss >= ''' + str(r['hits']) + ''' ?? Effect:|**//Threshold - [^tech.ss]//**^^''' + clean_effect + '\n'
    if t['skill'] == 'Vivus':
        res += '''--?? $tech.ones > 0 ?? Trigger:|The Cycle's Drain \n'''
    else:
        res += '''--?? $tech.ones > 0 ?? Trigger:|Recoil \n'''
    res += '''}}'''
    print(res)
    tech_pcards.append(res)
    # print(t['pcard'])
```
