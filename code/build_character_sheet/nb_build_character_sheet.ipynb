```python
df_talents = pd.DataFrame()
for i in df_talents_raw.index:
    df_talents = pd.concat([df_talents, pd.DataFrame(df_talents_raw.iloc[i, 0])], axis=0)
df_talents.reset_index(drop=True, inplace=True)
```

```python
df_talents
```

```python
rvr = df_talents[df_talents.tree == 'Reaver'].copy()
```

```python
rvr['tal_attr_nm'] = rvr.name.str.replace(' ', '_')
rvr['tal_attr_nm'] = rvr['tal_attr_nm'].str.lower()
rvr['path_attr_nm'] = rvr['tree'].str.lower()
rvr['div_id_nm'] = 'tal_' + rvr['type'].str.lower() + '_' + rvr['path_attr_nm'] + '_' + rvr['tal_attr_nm']
```

```python
rvr
```

```python
rvr.reset_index(drop=True, inplace=True)
```

```python
str_cbox = ""
for i in rvr.index:
    str_cbox += f'''<label for="{rvr.iloc[i]['div_id_nm']}">{rvr.iloc[i]['name']}</label>\n<input type="checkbox" id="{rvr.iloc[i]['div_id_nm']}">\n'''
```
