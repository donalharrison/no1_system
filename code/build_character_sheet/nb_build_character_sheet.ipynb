```python
def build_powercard(name, effect, skill, type):
    skill_clean = skill.lower()
    skill_clean = skill_clean.replace(' ', '_')
    effect_clean = effect.replace('choose', '**choose').replace(':<ul><li>', '**:^^^^**').replace('</li><li>', '^^^^').replace('</li></ul>', '**').replace('<strong>', '**').replace('</strong>', '**').replace('<em>', '//').replace('</em>', '//').replace('<br>', '^^').replace("<h4 style='margin: 3px;'><span style='background-color: #096491;border-radius: 8px'>", "**//").replace('</span></h4>', '//**^^')
    powercard = '''!power {{
    --name|@{character_name} - ''' + name + '''
    --Effect|''' + effect_clean + '''
    --Rank| d@{skill_''' + skill_clean + '''}
    --Result Set|[[ [$talent|XPND] @{Cosm|challenge}d@{skill_''' + skill_clean + '''}>@{Cosm|target}]]
    --Hits|[^talent.ss]
    --1s|[^talent.ones]
    --format|''' + type.lower() + '''check
    }}'''
    return(powercard)
```
